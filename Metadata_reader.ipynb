```python
#Reading all the json files in a folder vendor_metadata
import os
from os import listdir
from os.path import isfile, join
from shutil import copyfile

input_dir = "input//vendor_metadata//"
onlyfiles = [f for f in listdir(input_dir) if isfile(join(input_dir, f))]


#Creating dataframe for vendormetdata
vendor_metadata = pd.DataFrame()

for file in onlyfiles:
    data = json.load(open(input_dir + file ))
    for d in data:
        vendor_metadata =vendor_metadata.append(d,ignore_index=True)  
print("All the json files read into dataframe Successfully")        
```
